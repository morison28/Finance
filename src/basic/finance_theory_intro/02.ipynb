# 2. **貨幣の時間価値とキャッシュフローの評価**

## **2-1. キャッシュフロー**

　自分にとって受け取る資金を**キャッシュインフロー**，支払う資金を**キャッシュアウトフロー**と呼び，まとめて**キャッシュフロー**と呼ぶ．
決められた期日に，決められた額が約束どおりに発生するキャッシュフローを**確定的キャッシュフロー**と呼ぶ．
資金返済の約束は絶対で，帰ってこないことや半分になって帰ってくること(貸倒れ)，さらには遅れて返ってくること(延滞)は絶対にないことを前提にしている．
確定的キャッシュフローを**リスクのない**キャッシュフローということもある．
一方，事前に発生額を確定できないキャッシュフローを**確率的**キャッシュフローあるいは**不確実**キャッシュフローと呼ぶ(**リスクのある**キャッシュフロー)．

### **2-1-1. 確定的なキャッシュフロー**
　代表例は国債．債券とは発行体が広く世間から借金をする際に用いられる資金返済の約束を書いた証書．
返済の単位がその証書に記されており，それが**額面**．日本政府が発行する国債は，割引債でも利負債でも日本国債と呼び，
市場では現在までのところ，日本政府が国債を**償還**(満期に額面を返済すること)できないことはないと考えられているので，日本国債に関わるキャッシュフローは確定的と言える．

### **2-1-2. 不確実なキャッシュフロー**
　不確実なキャッシュフローの例としてはスワップがある．    
　**スワップ**とは，2つの経済主体が将来のキャッシュフローを交換する契約のこと．
例えば，**金利スワップ**では，同一通貨における変動金利と固定金利のキャッシュフローを交換する．
他に，異なる通貨間の金利と元本を交換する通貨スワップや金利と株価インデックスを交換するエクイティスワップなどが市場で取引されており，これらのスワップ取引で
世界のデリバティブ取引の大半を占めていると言われる．    
　金利スワップの標準的なタイプは，予め決められた元本(想定元本)に対して，半年ごとに固定金利による利息と変動金利による利息を交換するものである．
スワップ契約時点での利息交換と満期での元本交換は行われない．
変動金利サイドの指標金利としては通常LIBORレートが使用され，この金利を使って半年後の利払額が決まる．
一方，固定金利サイドでは，両者の経済的価値が等しくなるように計算された固定金利(これを**スワップレート**と呼ぶ)を支払う．    
　例えば，投資元本が100億円，満期4年，スワップレートが2％の金利スワップを考える．
固定サイドでは半年に1億円のりばらいが発生する．
一方，変動金利サイドの利払いは半年前のLIBORレートで決まるが，LIBORレートは不確実で，将来の値は誰にもわからない．
仮にLIBORレートが0.5年おきに，2.2 → 1.9 → 2.0 → 1.8 → 1.6 → 1.7 → 1.9 と変動した場合．半年後の変動金利サイドほ利払いは(現在のLIBORレートが2.2%なので)
1.1億円となり，実際の金利交換としては，変動サイドがら固定サイドへ差額の1000万円が支払われることになる．
同様に，1年後の変動サイドの利払いは(半年後のLIBORが1.9％なので)0.95億円となり，この場合には逆に，固定サイドから変動サイドへの差額の500万円が支払われることになる．
このような不確実な契約を結ぶ理由については第6章で触れる．

## **2-2. 割引債価格と利回り**

### **2-2-1. 単利と複利**

* 利回り：投資元本がどれくらいの率で利殖されたか
* 元利合計：投資元本 + 利息

額面P円，満期T年の割引債の発行価格がv円だったとすると，この割引債の利回りは，単利(年率表示)で，

$$
r_{s} = \left(\frac{P-v}{v}\right) \ / \  T
$$

* 単利：投資によって増加した金額が投資元本の何割にあたるかを年率で表したもの
* 1年複利：投資元本が1年ごとに再投資されることを考慮した場合の利回りで，利息が投資元本に加算されて，利息にも利息が付いてくる

額面P円，満期T年の割引債の1年複利利回りをrとすると，割引債の価格vは，

$$
v = \frac{P}{(1+r)^{T}}
$$

により求められる．逆に，これをrについて解くことで，1年複利利回りを

$$
r = \left(\frac{P}{v}\right)^{1/T} - 1
$$

として求めることだできる．割引債価格vと1年複利利回りは，一方がわかると他方が計算できる．

### **2-3-1. 転化回数**

　利負債では半年ごとに利息が支払われることが多いため，債券の実務では1年複利利回りではなく半年複利利回りが使用される．
**半年複利利回り**は，半年ごとに再投資した場合における，1年あたりの複利利回りを指す．    
　1年に何回再投資するかを**転化回数**と呼ぶ．半年複利では転化回数は2，4四半期複利では4である．
満期T年，額面P円，発行価格v円の割引債において，転化回数nの複利利回りrは年率表示で，

$$
v \times \left(1+\frac{r}{n}\right)^{nT} = P
$$

を満たす．    

　再投資の期間を短くした場合 ( 転化回数 $\rightarrow \infty$ )，$\frac{r}{n} = \frac{1}{N}$とすると$n \to \infty , N \to \infty$となり，
 
$$
\lim_{n \to \infty} \left(1+\frac{r}{n}\right)^{nT} = \lim_{N \to \infty} \left(1+\frac{1}{N}\right)^{rTN} 
= \lim_{N \to \infty} \left\{ \left(1+\frac{1}{N}\right)^{N}  \right\} ^{rT} = e^{rT} 
$$
を用いて，

$$
v \times e^{rT} = P
$$

で，rについて解くと，

$$
r = \frac{1}{T}\log{\left(\frac{P}{v}\right)}
$$

が得られる．連続複利利回りについても年率で表示されることに注意．    



### **2-3-2. 利回り曲線とフォワードレート**

　割引債の利回りは満期ごとに異なる値を通ることが多く，定期預金の利率も満期によって異なる．
いま，n年満期の割引債の1年複利利回り$r_{n}$が，
$$
r_{1}=1.5\%,r_{2}=2.0\%,r_{3}=2.5\%,r_{4}=3\%,r_{1}=3.5\%
$$
のように与えられたとする．
このような利回りと満期の組合せを**金利期間構造**あるいは**利回り曲線**と呼ぶ．    

　例えば，上記のような金利キアkん構造で，現在持っている100円を割引債で運用する．
次の2つの方法では，2年後に得られる運用結果に違いは生じるか

1. 満期2年の割引債に100円投資する
2. 満期1年の割引債に100円投資し，1年後に償還された元利合計の金額をその時(1年後)に発行される新しい満期1年の割引債に投資する

　1の方法では2年後に得られる元利合計は確定しており，$r_{2}=2.0\%$なので，
$$
100 \times (1 + 0.02)^{2} = 104.04
$$

となる．一方，2の方法では1年後の新しい満期1年の割引債の利回りは現時点ではわからないため，
投資した100円が2年後にいくらになるかは確定的ではない．
実際には，ほとんどの場合で1と2では異なる結果が得られる．
しかし，｢両者は同じ結果をもたらすはず｣とすると，このような考え方は**期待理論**と呼ぶ．
期待理論によれば，1年後の満期1年の割引サインの1年複利利回りを$f_{1,1}$とすれば，

$$
(1+r_{2})^{2} = (1+r_{1}) \times (1+f_{1,1})
$$

が成立するので，これから$f_{1,1}$が求まる．すなわち，期待理論を前提にすると，
将来の利回りについて情報を得たことになる．    
　同じ様に考えると，n年先の満期T年の割引債の1年複利利回り$f_{n,T}$も簡単に求まる．
$f_{n,T}$をn年s回T年**フォワードレート**(先渡利回り)と呼ぶ．
以降では，フォワードレートを1年複利利回り(年率表示)する．    
　$f_{n,T}$は満期n年の割引債と(n+T)年の割引債を用いて求めることができる．
満期n年の割引債の複利利回りを$r_{n}$，n年先T年フォワードレート$f_{n,T}$は，

$$
(1+r_{n})^{n}(1+f_{n,T})^{T} = (1+f_{n+T})^{n+T}
$$


### **2-2-4. 割引関数**

　額面1円の割引債の価格を満期T年の関数としてみたものを割引関数と呼び$d(T)$で表す．
満期T年の割引債の1年複利利回りを$r_{T}$とすると，割引関数$d(T)$は，

$$
d(T) = \frac{1}{(1+r_{T})^{T}}
$$

で定義される．割引関数は，利回り曲線から計算するか，もしくは割引債価格から直接計算する．

## **2-3. 割引現在価値と利負債価格**
　一般には現金と言うと紙幣やコインなどの物理的な対象を指し，きちんと保管してあれば時間が経過しても価値は変わらない．
しかし，ファイナンス理論においては，所定の利息と生む預金を指し，時間が経過するとその利息分だけ価値は｢必ず｣上昇する．   
　現金の価値が時間に関して増加した変化分を現金の**時間価値**と呼び，このことを｢現金には時間価値がある｣という．
時間価値により，現在の100円と1年後の100円の価値は異なる．
しかし．異なる時点でのキャッシュフローを単純に比較することはできない．
この不便さは，**割引現在価値**(Present Value; PV)で解消される．
割引現在価値は，将来のキャッシュフローに対し，対応する年限(満期)の割引関数をかけたものである．

### **2-3-1. キャッシュフローの分解**
　キャッシュフローXが以下で構成されているとする．

- 額面100円･満期1年
- 額面50円･満期2年
- 額面200円･満期3年
- 額面30円･満期4年

Xの現在価値の合計を**総割引現在価値**といい，計算すると，

$$
100 \times d(1) + 50 \times d(2) + 200 \times d(3) + 30 \times d(4) = 358.9551
$$

### **2-3-2. 利負債の価格**

 利負債は，満期までに決められた間隔で利息を受け取ることが出来る債券．
満期4年，額面100円，クーポンレート5％の利負債のキャッシュフローを考える．
この利負債の価格はいくらかを考える．
クーポンの部分をそれぞれ額面が5円の1，2，3，4年の割引債と考え，
残りを額面が100円で満期4年の割引債に分解して，その総割引現在価値を考えればいい．
　一般に，額面P円，クーポンが1年毎にC円，満期T年の利府祭の価格vは，割引関数が$d(i) \ (i=1,2,3,\cdot\cdot\cdot,T)$のとき，
 
$$
v = C\sum^{T}_{i=1} d(i) + P \times d(T)
$$

により与えられる．また割引関数と利回り曲線の関係から，利回り曲線が$r_{i}\ (i=1,2,3,\cdot\cdot\cdot,T)$のとき，利負債価格は

$$
v = \sum^{T}_{i=1} \frac{C}{(1+r_{i})^{i}} + \frac{P}{(1+r_{T})^{T}}
$$

と書くこともできる．


### **2-3-3. 最終利回り**

　投資した債券が平均的に何％で運用されたかを示す指標に**最終利回り**がある．
最終利回りは，キャッシュフローおよび投資に必要な価格から逆算された利回りで，次のように定義される．    

　額面P円，クーポンが1年毎にC円円，満期T年の利負債にv円を支払ったとすると，最終利回り(1年複利利回り)は，
 
$$
v = \sum^{T}_{i=1} \frac{C}{(1+y)^{i}} + \frac{P}{(1+y)^{T}}
$$

を満たすyで定義される．    

　最終利回りを考える際には，本来は満期により異なる利回りを一定値として考えていることに注意する．
上記の式をyについて解けば最終利回りが得られる．    

　債券取引の実務では，価格の代替として最終利回りを用いることが一般的．    
　クーポンレートと最終利回りが一致する時，利府祭の価格は額面に等しくなる．    
 
　国債市場や社債市場では新規発行される債券の価格は，額面に一致するようにクーポンレートで調整されており，このことを債券の**額面発行**あるいは**パー発行**という．
新たに発行された債券のことを**新発債**と呼び，新発債の価格は額面に一致する．    
　価格と額面が一致するような債券を発行しようとするならば，クーポンレートを最終利回りにすればよいことがわかる．
このとき，将来償還される額面と借入金額が一致するので，会計上の利息の額が明確になる．
一般に，価格が額面に等しい利負債を**パー債券**，額面以上の債券を**オーバーパー債券**，額面以下の債券を**アンダーパー債券**と呼ぶ．
また，価格が額面に等しい状態をパー，額面以下の状態をアンダーパー，額面以上の状態をオーバーパーと呼ぶ．

## **2-4. 投資の評価**

### **2-4-1. 内部収益率法**

　将来キャッシュフローが1年ごとに$C_{i} \ (i=1,2,3,\cdot\cdot\cdot,T)$で表され，
このキャッシュフローに対する投資の評価がvであるとき，内部収益率yは，

$$
v = \sum^{T}_{i=1} \frac{C_{i}}{(1+y)^{i}}
$$

により定義される．これにより，利負債だけでなく，任意のキャッシュフローの割安･割高を判断することができる．
キャッシュフローが確定的であれば，内部収益率の高い方がより合理的な投資と言える．

### **2-4-2. 正味現在価値法**

　内部収益率は将来のキャッシュフローと総割引現在価値から決定されるので，逆に，将来のキャッシュフローと利回りから総割引現在価値を求めることも出来る．
この様に予め定められた利回りを**割引率**という．割引率は利回り曲線を単純化のために一定化したものと考えられる．
割引率によってけいsなんされた将来キャッシュフロー$C_{i}$の総割引現在価値vから，その投資に関する実際の投資元本$C_{0}$を差し引いたものを
**正味現在価値(Net Present Valuel; NPV)**と呼ぶ．
すなわち，割引率をr，将来キャッシュフロー$C_{i}$の対価を$C_{0}$とすると，NPVは，

$$
NPV = -C_{0} + \sum^{T}_{i=1} \frac{C_{i}}{(1+r)^{i}}
$$

　正味現在価値NPVが正ならば，その投資は合理的であると考えてもう良いだろう．
逆にNPVが負になる投資は非合理的である．そのような評価方法を**正味現在価値法**．

## **2-5. リスクのあるキャッシュフロー**

　将来キャッシュフローが不確実な場合の投資評価方法はどうなるか．
一番簡単な方法は，不確実なキャッシュフローの平均値を使って計算すること．
数学的には，将来のキャッシュフローを確率変数で表し，その期待値を用いる．
このとき得られるキャッシュフローの対価をvとすると，内部収益率は，

$$
v = \sum^{T}_{i=1} \frac{E[C_{i}]}{(1+y)^{i}}
$$

で決定される．確率変数の期待値を計算するには確率分布をしらなくてはならないので，言うほど簡単じゃない．
NPVに関しては割愛．
