多次元の確率分布について

### **独立な確率変数**

　同時確率分布において，あらゆるx，yについて，条件

$$
f(x,y) = g(x)\cdots h(y)
$$

が成り立つならば，X,Yは**互いに独立**であるという．    

　独立とは関連が無いことであるが，同じ｢関連がない｣といっても，無相関よりはずっと強い．
無相関は平均的な性質であって確率分布から決まる変量$\rho$であるが，独立性は基礎の確率分布そのものに関する仮定だからである．    

　独立ならば無相関となる．しかし，無相関でも独立であるとは限らない．

### **多次元正規分布**

　2次元正規分布から考えてみる．$X_{1}, X_{2}$は独立で，N(0,1)に従う確率変数である．
また，a, b, c, d は定数として，確率変数の変換

$$
Y_{1} = aX_{1} + bX_{2} \\
Y_{2} = cX_{1} + dX_{2}
$$

を行なったとき，期待値，分散，共分散，相関係数，同時確率密度関数を求めてみる．    

　期待値は，$E(X_{1})=E(X_{2})=0$であるから，線形性により$E(Y_{1})=E(Y_{2})=0$となる．
よって，$\mu_{1}=\mu_{2}=0$．    


 分散は，独立性から分散の加法性が成立し，
$$
V(aX_{1}+bX_{2}) = a^{2}V(X_{1}) + b^{2}V(X_{2}) = a^{2} + b^{2}
$$

さらに独立性から$E(X_{1}X_{2})=E(X_{1})E(X_{2})=0$であるから，

$$
E(Y_{1}Y_{2}) = E((aX_{1} + bX_{2})(cX_{1} + dX_{2})) = E(acX_{1}^{2} + bdX_{2}^{2}) = acE(X_{1}^{2}) + bdE(X_{2}^{2}) = ac + bd
$$

つまり，分散，共分散，相関係数は，

$$
\sigma_{1}^{2} = a^{2} + b^{2}, \ \ \ \ \sigma_{2}^{2} = c^{2} + d^{2}, \ \ \ \ \sigma_{12} = ac + bd, \ \ \ \ \rho = \sigma_{12} / \sigma_{1}\sigma_{2}
$$

 同時確率密度関数については，まず$X_{1},X_{2}$はそれぞれN(0,1), N(0,1)に従うから，それらの同時確率分布は密度関数の積となって

$$
f(x_{1},x_{2}) = \frac{1}{\sqrt{2\pi}} \ exp \ (-\frac{x_{1}^{2} + x_{2}^{2}}{2})
$$

である．$Y_{1},Y_{2}$の同時確率密度関数$g(y_{1},y_{2})$ついては，変換

$$
\phi: y_{1} = ax_{1} + bx_{2}, \ \ \ \ y_{2} = cx_{1} + dx_{2}
$$

を逆に解いて，逆変換

$$
\psi: x_{1} = a'y_{1} + b'y_{2}, \ \ \ \ x_{2} = c'y_{1} + d'y_{2}
$$

ただし，

$$
a' = -\frac{d}{D}, \ \ \ b' = -\frac{b}{D}, \ \ \ c' = -\frac{c}{D}, \ \ \ d' = \frac{a}{D} \ \ \ (D=ad-bc)
$$

を，$f(x_{1},x_{2})$に代入すればよい．これで$g(y_{1},y_{2})$の主要部分が得られる．つまり，

$$
f(x_{1}, x_{2}) = f(a'y_{1}+b'y_{2}, c'y_{1}+d'y_{2})
$$

　密度関数は面積あたりのものであるから，変換$\psi$による**面積の伸縮率**をも考えるべきである．
4点$(y_{1},y_{2})=(0,0),(0,1),(1,0),(1,1)$のつくる面積1の単位正方形がxの空間に移された平行四辺形で，どれだけの面積を持つか調べれば，
平面の幾何学から，簡単に$|a'd' - b'c'|$とわかる．実際，$|a'd' - b'c'| = 1\|D|$である．    
　したがって，同時確率密度関数は，$f(x_{1},x_{2})$に代入して，

$$
g(y_{1},y_{2}) = f(a'y_{1} + b'y_{2}, c'y_{1} + d'y_{2})\cdot(1\|D|)
$$

で求められる．2乗和の部分を計算して，

$$
(a'y_{1} + b'y_{2})^{2} + (c'y_{1} + d'y_{2})^{2} = Ay_{1}^{2} + Cy_{1}y_{2} + By_{2}^{2}
$$

ただし，

$$
A = \frac{1}{\sigma_{1}^{2}(1-\rho^{2})}, \ \ \ B = \frac{1}{\sigma_{2}^{2}(1-\rho^{2})}, \ \ \ C = \frac{-2\rho}{\sigma_{1}\sigma_{2}(1-\rho_{2})}
$$

である．ここで，恒等式$(ac+bd)^{2}+(ad-bc)^{2} = (a^{2}+b^{2})(c^{2}+d^{2})$から，

$$
D^{2} = \sigma_{1}^{2}\sigma_{2}^{2} - \sigma_{12}^{2} = \sigma_{1}^{2}\sigma_{2}^{2}(1-\rho^{2})
$$

となることを用いた．結果として，同時確率密度関数は，

$$
g(y_{1},y_{2}) = \frac{1}{2\pi\sigma_{1}\sigma_{2}\sqrt{1-\rho^{2}}}
exp(
    -\frac{1}{2(1-\rho^{2})}
        (  \frac{y_{1}^{2}}{\sigma_{1}^{2}}
         - \frac{2\rho y_{1}y_{2}}{\sigma_{1}\sigma_{2}}
         + \frac{y_{2}^{2}}{\sigma_{2}^{2}}
        )
)
$$

となる．これを平均をそれぞれ0,0，分散をそれぞれ$\sigma_{1}^{2}, \sigma_{2}^{2}$，共分散を$\sigma_{12}$とする**2次元正規分布**といい，
$N((0,0), (\sigma_{1}^{2}, \sigma_{2}^{2}, \sigma_{12}))$と表す．なお，一般に平均$\mu_{1}, \mu_{2}$であるときには，
$y_{1}$が$y_{1}-\mu_{1}$に，$y_{2}$が$y_{2}-\mu_{2}$になる．$N((\mu_{1}, \mu_{2}), (\sigma_{1}^{2}, \sigma_{2}^{2}, \sigma_{12}))$

### **独立な確率変数の和**

#### **和の確率分布**

　離散型の確率変数X,Yが独立であるとし，その確率分布をg(x), h(y)としよう．
和X+Yの確率分布k(z)は確率P(X+Y=z)を考えれば得られる．
X+Y=zとなるのは，X=x, Y=z-xの形のx,z-xの加えてzになる全ての組合せであるから，それを確率の積で表して，

$$
k(z) = \sum_{x}g(x)h(z-x)
$$

となる．関数g,hからkを作る数学操作をg，hの畳み込みといい，k=g＊hとかく．g，hが密度関数のときも同様で，

$$
k(z) = \int_{-\infty}^{\infty} g(x) h(z-x) dx
$$
