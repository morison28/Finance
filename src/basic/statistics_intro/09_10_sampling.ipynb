## 09: 標本分布について

　**母集団**から**標本**を**抽出**し，母集団を推測する．
より具体的には**母集団分布**を推測する．
母集団分布が◯◯分布の形で与えられていれば**母数**を推定するだけでよい(**パラメトリック**)．
**ノンパラメトリック**の場合，広く定義できる大意表的なパラメータを使用し母集団分布を分析する．    

　標本平均のような，標本を要約し，母集団の母数の色々な推測に使われるものを**統計量**と呼ぶ．    

　統計量が，知りたい母集団の手がかりを与える．したがって，統計量の値がどのようなものかを知る必要性がある．母集団分布をf(x)としたとき，統計量の確率分布を**標本分布**と呼ぶ．
母集団と標本を結びつけるものとして，標本分布は統計的推測の考え方の中心になる概念である．    

　標本和や標本平均の具体的な標本分布は母集団分布に依存する．パラメトリックの場合においては，その分布が再生性を持っている際には，簡単に求めることができる．    
 
#### **二項母集団**    

母集団分布が母数pのベルヌーイ分布ならば二項分布Bi(1,p), $X_{1}+\cdot\cdot\cdot+X_{n}$の分布はBi(n,p)に従う．    

例) QC(Quality Control, 品質管理)     

　工場において製品が不良である確率はpであり，n個の製品に含まれる不良品の数の分布は二項分布Bi(n,p)である．
しかし，品質管理が厳重に行われている工業製品の母集団ではpはかなり小さいため，この場合はポアソン母集団と考えるべきであろう．    
　二項母集団が比較的よく当てはまるのは，生産過程の品質管理をあまり有効に行うことができない場合，
例えば果物などの農作物であろう．りんご，みかんなどの選定作業を考えてみると，p=0.1, n=10000の場合には，Bi(10000, 0.1)に従う．
基準をややゆるくしてp=0.05にした場合にはBi(10000,0.05)が当てはまる．この場合には，nが大きいので，次際に二項分布を用いて計算することはなく，
中心極限定理から近似として正規分布に持ち込むことが多い．

#### **ポアソン母集団**

　母集団分布が母数$\lambda$のポアソン分布$Po(\lambda)$に従うならば標本和は$Po(n\lambda)$に従う．    

例)交通事故死亡者数     

　日本全国で，1日当たりの交通事故による死亡者数の平均が30人である場合，1年の死亡者の分布はPo(10950)である．
ただし，実際の計算上は，Po(30)を当てはめる分布としては現実的ではない．なぜなら，ポアソンにおける$e^{-30}$の計算は必ずしも現実ではない($e^{-10950}$はなおさら)．
実際には，$\lambda$の大きさの限界として1桁のオーダーでない限り，中心極限定理による他ないからである．そのためには，｢日本全国｣を｢県｣｢市｣あるいは｢警察署管内｣くらいに小さくするべきである．また，一定人口数あたり10万人当たりなどで考えることも同じ目的に添う．

#### **正規母集団**

　母集団正規分布$N(\mu, \sigma^{2})$ならば，標本和は$N(n\mu, n\sigma^{2})$に従い，標本平均は$N(\mu, \frac{\sigma^{2}}{n})$に従う．    
 
例)測定誤差     
　
 100cmの棒の長さを測定すると，1回ごとの測定は測定誤差のため平均100cm，標準偏差0.1cmの正規分布(100, 0.1^2)に従うとする．
この測定を100回繰り返してその平均を考えると，その分布，平均100cm，標準偏差0.01cmの正規分布N(100, 0.01^2)に従う．    
　正規母集団がよく当てはまる例としては同一企画のものをn個製造ないしは測定する場合がある．
たとえばコインの質量があるが，いくつかの経験から，これらの計測値の分布は正規分布であることが多い．

#### **漸近的正規性**

　一般に標本分布を求める場合は重積分(畳み込み)が必要であり，正確に求めるのは困難であったり不可能である場合も多い，
しかしながら，このような場合でも中心極限定理によって近似的にその分布を求めることができる．
すなわち，nが十分大きければ(たとえば，30以上)，母集団分布に関係なく，標本平均の分布は期待値$\mu$，分散$\sigma^{2}/n$の正規分布であるから，

$$
P(a \le \frac{\bar{X}-\mu}{\sigma/\sqrt{n}} \le b) \fallingdotseq \int_{a}^{b} \frac{1}{\sqrt{2\pi}} e^{-\frac{x^{2}}{2}} dx = \Phi(b) - \Phi(a)
$$

のように近似できる．    

　なお，$n \rightarrow \infty$の場合に成り立つこと(nが十分に大きい場合，近似的に成り立つこと)を統計学では漸近的と言う用語で表す．
したがって，上述したことは｢標本平均$\bar{X}$の漸近分布が平均$\mu$，分散$\sigma^{2}/n$の正規分布$N(\mu, \sigma^{2}/n)$である，と言い換えられる．

## 10: 正規分布からの標本

　｢正規母集団｣とは，母集団分布が正規分布であるような母集団である．
正規母集団から標本がとられるという仮定も，統計学の理論の中心である．
このとき，正規母集団からの標本$X_{1}, \cdot\cdot\cdot, X_{n}$に基づく統計量の標本分布を計算することになるが，
この計算を行うのが**正規標本論**である．    

　統計分析を行う際にに正規分布を仮定するのは以下の理由がある．
- 身長の分布や実験誤差の分布など我々が分析対象とする現象は正規分布で表せるものが多い
- 体重や所得の分布の様に変数の変換によって(体重は1/3乗し，所得は対数をとる)正規分布で表せることが多い
- 標本平均などの多くの統計量は，確率変数の和になっているものが多い．このような統計量の標本分布は中心極限定理によって漸近的に正規分布であり，その標本分布は，
正規母集団の場合の結果を近似的にそのまま使うことができる．
- 独立な正規確率変数の和は正規確率変数であるなど数学的な扱いが容易であり，標本分布を正確に求めることができる．



例)株価     

　確率変数Xはある株式の利回りで(%)で，平均3，分散10の正規分布N(3,10)に従うものである．株式投資が損となる確率を求めてみる．    
　$\sigma=\sqrt{10}=3.162$であるから，条件$X<0$は$Z<-0.949$となり，$P(X<0)=P(Z<-0.949)=1-\Phi(0.949)=0.171$である．
